```python
import pandas as pd
import codecs
import re

# df = pd.read_json(codecs.open('DataJSON/sentiment_4_manh_210.json','rU','UTF-8'))
# df = pd.read_json(codecs.open('DataJSON/sentiment_4_manh_214.json','rU','UTF-8'))
# df = pd.read_json(codecs.open('DataJSON/sentiment_4_manh_309.json','rU','UTF-8'))
# df = pd.read_json(codecs.open('DataJSON/sentiment_4_manh_676.json','rU','UTF-8'))
df = pd.read_json(codecs.open('DataJSON/sentiment_4_manh_1493.json','rU','UTF-8'))
```

```python
# Remove emoji
emoji_pattern = re.compile("([\u263a-\U0001ffff])", flags=re.UNICODE)
df.content = df.content.str.replace(emoji_pattern, '')
```

```python
# Remove content in ()
pattern = re.compile("(\(.*?\))|(\[.*?\]|(<.*?>))", flags=re.UNICODE)
df.content = df.content.str.replace(pattern, '')
```

```python
pattern = re.compile("\\n\\n", flags=re.UNICODE)
df.content = df.content.str.replace(pattern, '. ')
pattern = re.compile("(\.\.+)", flags=re.UNICODE)
df.content = df.content.str.replace(pattern, '. ')
```

```python
pattern = re.compile("(<.*?>)")
df.content = df.content.str.replace(pattern, '')
pattern = re.compile("(</.*?>)")
df.content = df.content.str.replace(pattern, '')

pattern = re.compile("(\#[0-9\W]+|\\\"|(?:[0-9]+\.)+[0-9]{0,61}|[0-9]+\.[0-9]+[a-zA-Z]+|[0-9]+\.[0-9]+)")
df.content = df.content.str.replace(pattern, '')

pattern = re.compile("http[s]?://((?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+)[^\.\s]")
df.content = df.content.str.replace(pattern, '')


pattern = re.compile("(?:[a-z0-9](?:[a-z0-9-]{0,61}[a-z0-9])?\.)+[a-z0-9][a-z0-9-]{0,61}[a-z0-9]")
df.content = df.content.str.replace(pattern, '')

pattern = re.compile("\\n")
df.content = df.content.str.replace(pattern, '.')

pattern = re.compile("TP\.")
df.content = df.content.str.replace(pattern, 'TP_')
```
