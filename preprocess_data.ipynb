```python
stopwords = []
with open('vietnamese-stopwords-dash.txt', 'r',encoding="utf-8") as f:
    for word in f:
        word = word.split('\n')
        stopwords.append(word[0])
```

```python
    import pandas as pd
    import codecs
    from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'[^\W0-9]+')
df = pd.DataFrame()
df.to_csv("data.csv", sep=',', encoding='utf-8', mode='w',header=False)
```

```python
for file in glob.glob("./Data/*.csv"):
    print(file)
    d = pd.read_csv(codecs.open(file,'rU','UTF-16'), sep='\t',header=0)
    d['tokenized']=''
    for index, row in d.iterrows():
        try:
            if row['SegmentedPost']=='' or row['SegmentedPost']==pd.np.nan:
                d.drop(index)
                continue
            d.at[index,'tokenized']=[t for t in tokenizer.tokenize(row['SegmentedPost']) if t not in stopwords]
        except:
            print(index)
            print(row["SegmentedPost"])
            d.drop(index)
    d = d[d["SegmentedPostLabel"].isin([0,1,-1])]
    d.to_csv("data.csv", sep=',', encoding='utf-8', mode='a',header=False)
    print(d.shape)
# df = pd.concat(list_, axis = 0, ignore_index = True)
# print(df.shape)
```
